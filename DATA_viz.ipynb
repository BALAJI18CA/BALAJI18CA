```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,'r*')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('BAL')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes1.set_xlabel("x")
axes1.set_ylabel('y')
axes1.set_title("BMW")
#axes2
axes2.plot(x,y,)
axes2.set_xlabel("x")
axes2.set_ylabel('y')
axes2.set_title("bmws")


fig
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes1.set_xlabel("x")
axes1.set_ylabel('y')
axes1.set_title("BMW")
#axes2
axes2.plot(x,y,)
axes2.set_xlabel("x")
axes2.set_ylabel('y')
axes2.set_title("bmws")
axes2.set_xlim(30,50)
axes2.set_ylim(20,30)
axes2.set_label('zoom')



fig
```

```python
fig,axes4= plt.subplots(nrows=1, ncols=2)
axes4[0].plot(x,y,color='blue',lw=3,ls='--')
axes4[1].plot(x,z,color='r',lw=3,ls='-')
fig
```

```python
axes4[0].plot(x,y,color="blue", lw=3, ls='--')
axes4[1].plot(x,z,color="red", lw=3, ls='-')
fig
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,2)) # size of the subplots
axes[0].plot(x,y,"y",lw=2,ls="-.",label="y--")
axes[0].set_xlabel("x")
axes[0].set_ylabel('y')
axes[0].set_title("bal")
axes[0].legend(loc=2)
#axes[1]
axes[1].plot(x,y,"r",lw=2,ls=":",label="x--")
axes[1].set_xlabel("x")
axes[1].set_ylabel('y')
axes[1].set_title("bal")
axes[1].legend(loc=2)
```
